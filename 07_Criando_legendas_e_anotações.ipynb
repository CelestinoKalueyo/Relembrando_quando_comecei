# Seção 7 - Criando legendas e anotações

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns 

%matplotlib inline
rcParams["figure.figsize"] = 22, 8
sns.set_style("whitegrid")

caminho = 'Downloads/arquivos_de_exercicios_python_para_ciencia_dados_basico/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ["nomes", "mpg", "cyl", "disp", "hpt", "drat", "wt", "qsec", "vs", "am", "gear", "carb"]
mpg = carros["mpg"]
```

# Títulos

```python
x = range(1, 10)
y = [1, 2, 3, 4, .5, 4, 3, 2, 1]
plt.bar(x, y)
plt.xlabel("Título no eixo x")
plt.ylabel("Título no eixo y")
```

# Legenda no gráfico

```python
import matplotlib.patches as mpatches

plt.plot(mpg)
plt.xlabel("nomes dos carros")
plt.ylabel("milhas/galão")
plt.title("Consumo de gasolina")
plt.xticks(mpg.index, carros.nomes, rotation=45)

legenda = mpatches.Patch(label="mpg")
plt.legend(handles=[legenda])
```
